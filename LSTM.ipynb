```python
import tensorflow as tf
import os
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/forecasting_dataset.csv')
```

```python
df
```

```python
df = df[5::6]
df
```

```python
df.index = pd.to_datetime(df['Date Time'], format='%d.%m.%Y %H:%M:%S')
df[:26]
```

```python
temp = df['T (degC)']
temp.plot()
```

```python
def df_to_X_y(df, window_size=5):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [[a] for a in df_as_np[i:i+window_size]]
    X.append(row)
    label = df_as_np[i+window_size]
    y.append(label)

  return np.array(X), np.array(y)
```

```python
WINDOW_SIZE = 5
X1, y1 = df_to_X_y(temp, WINDOW_SIZE)
X1.shape, y1.shape
```

```python
X_train1, y_train1 = X1[:60000], y1[:60000]
X_val1, y_val1 = X1[60000:65000], y1[60000:65000]
X_test1, y_test1 = X1[65000:], y1[65000:]
X_train1.shape, y_train1.shape, X_val1.shape, y_val1.shape, X_test1.shape, y_test1.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import *
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
from tensorflow.keras.optimizers import Adam

model1 = Sequential()
model1.add(InputLayer((5, 1)))
model1.add(LSTM(64))
model1.add(Dense(8, 'relu'))
model1.add(Dense(1, 'linear'))

model1.summary()
```

```python
cp1 = ModelCheckpoint('/content/model1.keras', save_best_only=True)
model1.compile(loss=MeanSquaredError(), optimizer=Adam(learning_rate=0.0001), metrics=[RootMeanSquaredError()])
```

```python
model1.fit(X_train1, y_train1, validation_data=(X_val1, y_val1), epochs=5, callbacks=[cp1])
```

```python
from tensorflow.keras.models import load_model
model1 = load_model('/content/model1.keras')
```

```python
train_predict1 = model1.predict(X_train1).flatten()
train_results = pd.DataFrame(data={'Train Predictions': train_predict1, 'Actuals': y_train1.flatten()})
train_results.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(train_results['Train Predictions'][50:100])
plt.plot(train_results['Actuals'][50:100])
plt.legend(['Train Predictions', 'Actuals'])
plt.show()
```

```python
val_predict1 = model1.predict(X_val1).flatten()
val_results = pd.DataFrame(data={'Validation Predictions': val_predict1, 'Actuals': y_val1.flatten()})
val_results.head()
```

```python
plt.plot(val_results['Validation Predictions'][50:100])
plt.plot(val_results['Actuals'][50:100])
plt.legend(['Validation Predictions', 'Actuals'])
plt.show()
```

```python
test_predict1 = model1.predict(X_test1).flatten()
test_results = pd.DataFrame(data={'Test Predictions': test_predict1, 'Actuals': y_test1.flatten()})
test_results.head()
```

```python
plt.plot(test_results['Test Predictions'][50:100])
plt.plot(test_results['Actuals'][50:100])
plt.legend(['Test Predictions', 'Actuals'])
plt.show()
```
